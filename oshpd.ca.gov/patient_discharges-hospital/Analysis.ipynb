```python
import os
from ambry.bundle import BuildBundle
b = BuildBundle(os.getcwd())

```

```python
df = b.partitions.find(table ='pdd_summary_mdc').pandas
dfmi = df.set_index(['oshpd_id','year','mdc'])
dfmi = dfmi.drop('id',1).unstack().fillna(0)
dfmi.to_csv('count_by_mdc.csv')
```

```python
p = b.library.dep('pddpuf').partition

q = "SELECT id, oshpd_id, dschyear as year,  msdrg FROM pdd_puf_c WHERE mdc = '14'"

df = p.select(q).pandas
dfx = df.groupby(['oshpd_id','year','msdrg']).count().unstack().fillna(0)
dfx
```

```python
dfr = dfx.reset_index()
dfr.columns =  dfr.columns.droplevel()
dfr.columns = ['oshpd_id','year'] + list(dfr.columns)[2:]
dfr.head()
```
